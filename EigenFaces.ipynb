```python
import PIL
from matplotlib import pyplot as plt
from matplotlib.image import imread
import matplotlib.cm as cm
import numpy as np
import os
```

```python
data_path = '/Users/amarkr/Documents/GitHub/ComputerVision/dataset/'
```

```python
#change the directory to the datapath
os.chdir(data_path)
def display_image(name,display=0):
    img = PIL.Image.open(name)
    gray_img = np.asarray(img.convert("L"))
    #print(clr)
    if(display==1):
        plt.imshow(gray_img,cmap=cm.gray)
        plt.show()
    return gray_img
    
```
